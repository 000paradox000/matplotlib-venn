```python
from matplotlib_venn import venn3
from tests.utils import verify_diagram
from matplotlib.pyplot import figure, subplot
```

```python
# The usual case
v = venn3((1, 1, 1, 1, 1, 1, 1))
test_points = {'111': [(0, 0)],
               '110': [(0, 0.3)],
               '101': [(-0.3, -0.15)],
               '011': [(0.3, -0.15)],
               '100': [(-0.3, 0.2)],
               '010': [(0.3, 0.2)],
               '001': [(0, -0.3)],
               '': [(0,0.5), (-0.42, -0.3), (0.42, -0.3)]}
verify_diagram(v, test_points)
```
